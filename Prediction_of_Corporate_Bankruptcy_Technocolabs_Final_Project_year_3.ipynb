METHOD 1 knn Data Imputation -MISSING VALUE

```python
d1= data.copy()
```

```python

from sklearn.impute import KNNImputer

# Replace '?' with NaN in the DataFrame
d1.replace('?', np.nan, inplace=True)

# Assuming 'data_keys' contains the list of column names you want to impute
for i in data_keys:
    X = np.array(d1[i]).reshape(-1, 1)

    # Convert the column to float (including NaNs) before imputing
    X = X.astype(float)

    imputer = KNNImputer(n_neighbors=3)
    X_prep = imputer.fit_transform(X)

    # Assign the imputed values back to the DataFrame
    d1[i] = X_prep.reshape(1, -1)[0]

d1.head()
```

```python

total_cells= np.product(d1.shape)
total_missing=d1.isnull().sum().sum()
# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

There is imbalance in the Target value, we need to oversample it based

OverSamppling :SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python

X=d1.drop(labels=['class'], axis=1)
y=d1['class']
```

```python
oversample = SMOTE()
X,y=oversample.fit_resample(X,y)
```
